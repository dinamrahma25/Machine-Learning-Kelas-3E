```python
from google.colab import drive
from google.colab.patches import cv2_imshow

image = '/content/drive/MyDrive/uts/dummy/26-E-4016-TO-08-22_jpeg.rf.8c72fd54437debb496cef33f54cfd44a.jpg'

# Membaca gambar
img = cv2.imread(image)

# Menampilkan gambar
cv2_imshow(img)
```

```python
h, w, ch= img.shape
print(h)
print(w)
print(ch)
```

```python
data = img / 255.0
data = data.reshape(h * w, ch)
data.shape
```

```python
def plot_pixels(data, title, colors=None, N=10000):
    if colors is None:
        colors = data

    # choose a random subset
    rng = np.random.RandomState(0)
    i = rng.permutation(data.shape[0])[:N]
    colors = colors[i]
    R, G, B = data[i].T

    fig, ax = plt.subplots(1, 2, figsize=(8, 4))
    ax[0].scatter(R, G, color=colors, marker='.')
    ax[0].set(xlabel='Red', ylabel='Green', xlim=(0, 1), ylim=(0, 1))

    ax[1].scatter(R, B, color=colors, marker='.')
    ax[1].set(xlabel='Red', ylabel='Blue', xlim=(0, 1), ylim=(0, 1))

    fig.suptitle(title, size=14)
```

```python
plot_pixels(data, title='Input color space: 16 million possible colors')
```

```python
from sklearn.cluster import KMeans

citra_reshape = img.reshape((-1, 3))

jml_klaster = 2

kmeans = KMeans(n_clusters=jml_klaster)
kmeans.fit(citra_reshape)

label_klaster = kmeans.labels_.reshape(img.shape[:2])

citra_segm = img.copy()
citra_segm[label_klaster == 0] = [0, 0, 0]
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

img_contrast = cv2.equalizeHist(img_gray)

citra_segm = cv2.imread('segmentation_image.jpg')

fig, ax = plt.subplots(1, 2, figsize=(16, 6), subplot_kw=dict(xticks=[], yticks=[], aspect='equal'))
fig.subplots_adjust(wspace=0.05)

ax[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
ax[0].set_title('Original Image', size=16)

ax[1].imshow(cv2.cvtColor(img_contrast, cv2.COLOR_GRAY2RGB))
ax[1].set_title('Equalized Image', size=16)

plt.show()
```
