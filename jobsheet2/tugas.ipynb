```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Membaca data
data = pd.read_csv('wbc.csv')
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
x = data.drop(["diagnosis", "radius_mean", "texture_mean", "perimeter_mean", "area_mean", "smoothness_mean", "compactness_mean",
         "concavity_mean", "concave points_mean", "symmetry_mean", "fractal_dimension_mean"], axis=1)
y = data['diagnosis']
data.head()
```

```python
le = LabelEncoder()
y = le.fit_transform(y)
data.head()
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
data['radius_mean'] = scaler.fit_transform(data[['radius_mean']])
data['texture_mean'] = scaler.fit_transform(data[['texture_mean']])
data['perimeter_mean'] = scaler.fit_transform(data[['perimeter_mean']])
data['area_mean'] = scaler.fit_transform(data[['area_mean']])
data['smoothness_mean'] = scaler.fit_transform(data[['smoothness_mean']])
data['compactness_mean'] = scaler.fit_transform(data[['compactness_mean']])
data['concavity_mean'] = scaler.fit_transform(data[['concavity_mean']])
data['concave points_mean'] = scaler.fit_transform(data[['concave points_mean']])
data['symmetry_mean'] = scaler.fit_transform(data[['symmetry_mean']])
data['fractal_dimension_mean'] = scaler.fit_transform(data[['fractal_dimension_mean']])
data.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, stratify=y, random_state=42)
print(f'Jumlah label data asli:\n{pd.Series(y).value_counts()}')
print(f'Jumlah label data train:\n{pd.Series(y_train).value_counts()}')
print(f'Jumlah label data test:\n{pd.Series(y_test).value_counts()}')
```
